```python
import requests
from bs4 import BeautifulSoup

tony_year_list = []
tony_title_list = []
tony_winner_list = []

import numpy as np

for tony_year in np.arange(1947, 2020):
    year = requests.get('https://www.tonyawards.com/winners/year/' + str(tony_year) + '/category/any/show/any/')
    year_soup = BeautifulSoup(year.text)
    
    tony_amount = year_soup.find_all('h2', class_='category_title')
    
    for winner_block in np.arange(0,len(tony_amount)):
        tony = year_soup.find_all('h2', class_='category_title')[winner_block]
        tony = str(tony).split('>')[1]
        tony = tony.split('<')[0]
        tony = tony.split('/')
        tony[0] = tony[0].strip()
        tony[1] = tony[1].strip()
        tony_year = tony[0]
        tony_title = tony[1]
        tony_year_list.append(tony_year)
        tony_title_list.append(tony_title)

        winner = year_soup.find_all('div', class_='inner-r')[winner_block]
        winner = str(winner).split('<h3>')[1]
        winner = winner.split('</h3>')[0]
        if tony_year == 2019: 
            tony_winner_list.append(winner)
        else:
            if winner[0:3] == '<i>':
                winner = winner.replace('<i>','')
                winner = winner.replace('</i>','')
                winner = winner.split('.')[0]
                winner = winner.split(',')[0]
                tony_winner_list.append(winner)
            else: 
                winner = winner.split('.')[0]
                winner = winner.split(',')[0]
                tony_winner_list.append(winner)                
        
tony_winner_list
```
